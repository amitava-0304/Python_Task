# 1 . load this data in pandas

```python
import pandas as pd
import numpy as np
dict_df = pd.read_excel('c://file//Superstore_USA.xlsx',['Orders','Returns','Users'])
```

```python
Orders = dict_df .get('Orders')
Returns = dict_df.get('Returns')
Users = dict_df.get('Users')
```

```python
Orders['Diff_Days'] = ((Orders['Ship Date'] - Orders['Order Date'])/np.timedelta64(1, 'D'))
Orders['Diff_Days'] = Orders['Diff_Days'].astype(int)
Orders['Profit_Margin'] = (Orders.Profit / Orders.Sales) * 100
```

# 15 . try to find out the 10th most loss making product catagory . 

```python
Orders['Profit_Margin'] = (Orders.Profit / Orders.Sales) * 100
```

```python
loss_making_product_catagory_df = pd.DataFrame(Orders.groupby(['Product Category'])[['Profit_Margin']].sum().
                                               sort_values('Profit_Margin',ascending = True).reset_index())
top10 = loss_making_product_catagory_df [-1:]
top10
```

# 16 . Try to find out 10 top  product with highest margins 

```python
product_highest_margins_df = pd.DataFrame(Orders.groupby(['Product Name'])
                                        [['Profit_Margin']].sum().sort_values('Profit_Margin',ascending = False).reset_index())
top10 = product_highest_margins_df[:10]
top10
```
