# 9.average weekly response time for each agent

```python
def get_sec(time_str):
    """Get seconds from time."""
    h, m, s = time_str.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)
```

```python

Agent_Performance['Response_time']=Agent_Performance['Average Response Time'].apply(get_sec)
Agent_Performance2 = Agent_Performance[Agent_Performance['Response_time'] > 0.00]
```

```python
Agent_Time = Agent_Performance2.groupby(['week_number','Agent Name'])['Response_time'].mean()/60
Agent_Time = Agent_Time.to_frame()
Agent_Time
```

# 13.Total contributation hour for each and every agents weekly basis

```python
Agent_Login = pd.read_csv('c:\\file\\Agent Login Report.csv',skiprows=2)
```

```python
Agent_Login['Date'] = pd.to_datetime(Agent_Login['Date'])
Agent_Login['Week'] = Agent_Login['Date'].dt.week
Agent_Login
```

```python
def get_hr(time_str):
    """Get hr. from time."""
    h, m, s = time_str.split(':')
    return (int(h) * 3600 + int(m) * 60 + int(s))/3600
```

```python
Agent_Login['Contributation hour']=Agent_Login['Duration'].apply(get_hr)
Agent_Login
```

```python
Agent_Time = Agent_Login.groupby(['Week','Agent'])['Contributation hour'].sum()
Agent_Time = Agent_Time.to_frame()
Agent_Time
```

# 14. total percentage of active hour for a month


```python
active_hour = ((Agent_Login.groupby('Agent')['Contributation hour'].sum()) / (30*24)) * 100
active_hour = active_hour.to_frame('Active hours percentage')
active_hour
```
