# 9. Try to find out total dress sell for individual dress id 

```python
def create_view(mydb, query):
    cursor = mydb.cursor()
    try:
        cursor.execute(query)
        print("View created successfully")
    except Error as err:
        print(f"Error: '{err}'")
```

```python
query="create view sales_view as select Dress_ID,sum(`29-08-2013`+`31-08-2013`+`09-02-2013`+`09-04-2013`+\
`09-06-2013`+`09-08-2013`+`09-10-2013`+`09-12-2013`+14-09-2013`+`16-09-2013`+`18-09-2013`+`20-09-2013`+\
`22-09-2013`+`24-09-2013`+`26-09-2013`+`28-09-2013`+`30-09-2013`+\
`10-02-2013`+`10-04-2013`+`10-06-2013`+`10-08-2010`+`10-10-2013`+`10-12-2013`)\
as Total_Sales from dress_sales group by Dress_ID;"
```

```python
mydb=database.create_database_connection()
create_view(mydb, query)
```
